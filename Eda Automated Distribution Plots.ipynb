```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
def graph_v1(file):
    dat=pd.read_csv(file)
    for i in dat.columns:
        if i in dat.loc[:,dat.dtypes==object]:
            dat[i].value_counts().plot(kind='bar', figsize=(10,8), color='turquoise', fontsize=12, rot=0)
            plt.xlabel(i,fontsize=15)
            plt.ylabel('Frequency',fontsize=15)
            fig=plt.gcf()
            plt.show()
            fig.savefig(i+'.jpg',dpi=100)
        else:
            plt.subplot(2,1,2)
            plt.hist(dat[i],histtype='bar',bins=20,color='turquoise',edgecolor='white')
            plt.xlabel(i)
            plt.ylabel('Frequency')

            plt.subplot(2,1,1)
            plt.boxplot(dat[i], vert=False)
            plt.title(f'Distribution of {i}')
            fig1=plt.gcf()
            plt.show()
            fig1.savefig(i+'.jpg',dpi=100)
```

```python
graph_v1('attrition.csv')
```

```python
def graph_time_series_v5(file, columns=1, dir=r'C:\study material\PYTHON\python final assignment', date_column='Date'):
    dat=pd.read_csv(file, index_col=date_column, parse_dates=True)
    import os
    os.chdir(dir)
    if columns==1:
        columns=dat.columns
    for i in columns:
        if i in dat.loc[:,dat.dtypes==object] or dat[i].value_counts().count()<10:
            dat[i].value_counts().plot(kind='bar', figsize=(10,8), color='turquoise', fontsize=12, rot=0)
            plt.xlabel(i,fontsize=15)
            plt.ylabel('Frequency',fontsize=15)
            fig=plt.gcf()
            plt.show()
            fig.savefig(i+'.jpg')
        else:
            dat[i].plot(figsize=(10,8))
            plt.title(f'Distribution of {i} across timeline',fontsize=20)
            plt.ylabel(i)
            fig=plt.gcf()
            plt.show()
            fig.savefig(i+'.jpg')
```

```python
def graph_time_series_v6(file,columns=1,dir=r'C:\study material\PYTHON\python final assignment',date_column='Date',date_format="D"):
    #here date_column is date column of time series against which graph is to be plotted
    #date format as 'D' for daily agg "M" for monthly agg similarly 'Y' for annual agg
    dat=pd.read_csv(file, index_col=date_column, parse_dates=True)
    import os
    os.chdir(dir)
    if columns==1:
        columns=dat.columns
    for i in columns:
        if i in dat.loc[:,dat.dtypes==object] or dat[i].value_counts().count()<10:
            dat[i].value_counts().plot(kind='bar', figsize=(10,8), color='turquoise', fontsize=12, rot=0)
            plt.xlabel(i,fontsize=15)
            plt.ylabel('Frequency',fontsize=15)
            fig=plt.gcf()
            plt.show()
            fig.savefig(i+'.jpg')
        else:
            figtrend=plt.figure(figsize=(15,15))
            plt.plot(dat[i].resample(date_format).sum(),c='blue')
            import statsmodels.api as sm
            a=sm.tsa.seasonal_decompose(dat[i].resample(date_format).agg(['sum']),model='additive')
            plt.plot(a.trend.index, a.trend, c='red')
            plt.xlabel(date_column,fontsize=15)
            plt.ylabel(i,fontsize=15)
            plt.title(f'Distribution of agg {i} across timeline',fontsize=20)
            fig=plt.gcf()
            plt.show()
            fig.savefig(i+'.jpg')
```

```python
graph_time_series_v6('sales_train.csv', columns=['item_cnt_day','item_price'], date_column='date')
```
